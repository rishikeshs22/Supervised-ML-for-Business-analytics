```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score


```

```python
df = pd.read_csv('CarPrice_Assignment (1) (3).csv')
df.head()
```

```python
#We need only certain features i.e, carlength, carwidth, carheight, enginesize,stroke,horsepower,price
filtered_data = df[['carlength','carwidth','carheight','enginesize','stroke','horsepower','price']]
filtered_data.head()
```

```python
train_filtered_data, test_filtered_data = train_test_split(filtered_data, test_size=0.2, random_state=42)
X_train = train_filtered_data.drop('price', axis=1)
Y_train = train_filtered_data['price']
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
#using r2_score to get accuracy(r squared score lies between 0 and 1. 1 represent perfect prediction)
X_test = test_filtered_data.drop('price', axis=1)
Y_test = test_filtered_data['price']
Y_pred = model.predict(X_test)
accuracy = r2_score(Y_test, Y_pred)

print("Accuracy of linear regression model:", accuracy)

```

```python
#to predict the price of car of A
features_A = np.array([[190.9, 70.3, 54.9, 183, 3.64, 123]])
predictedprice = model.predict(features_A )
print("Predicted price for Person A's car:", predictedprice[0])
```
